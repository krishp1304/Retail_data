```python
city_state_pairs = df.dropna(subset=['City', 'State'])[['State', 'City']]
```

```python
state_to_cities = city_state_pairs.groupby('State')['City'].unique().apply(list).to_dict()
```

```python
state_to_cities
```

```python
import numpy as np
```

```python
for state, cities in state_to_cities.items():
    df["State"] = np.where(df["State"].isnull() & df["City"].isin(cities),state,df["State"])
```

```python
df['Total_Purchases'] = np.where(df['Total_Purchases'].isnull() & (df['Amount'] != 0),df['Total_Amount'] / df['Amount'],df['Total_Purchases'])
```

```python
df['Amount'] = np.where(df['Amount'].isnull() & (df['Total_Purchases'] != 0),df['Total_Amount'] / df['Total_Purchases'],df['Amount'])
```

```python
df['Total_Amount'] = np.where(df['Total_Amount'].isnull() & df['Amount'].notnull() & df['Total_Purchases'].notnull(),df['Amount'] * df['Total_Purchases'],df['Total_Amount'])
```
